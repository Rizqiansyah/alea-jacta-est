```python
import numpy as np  
import matplotlib.pyplot as plt
from aje.special import gammainv
from scipy.special import gamma
from aje.special.gamma import alpha
```

```python
# Check the error

first_x = np.linspace(0.1, alpha, 1000)
second_x = np.linspace(alpha, 5., 1000)
first_y = gamma(first_x)
second_y = gamma(second_x)
first_soln = gammainv(first_y)[0]
second_soln = gammainv(second_y)[1]

plt.plot(first_x, first_x-first_soln)
plt.plot(second_x, second_x-second_soln)
```
